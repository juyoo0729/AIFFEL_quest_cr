사각형 넓이 구하기

```python
class Square:
    def __init__(self):
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')

        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')

        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')

        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    def par(self):
        base, hight = map(int, input('밑변, 높이를 입력하세요. 예시 : 밑변,높이\n >>>').split(','))
        area = base*hight
        result = '평행사변형의 넓이는 : ' + str(area)
        # [[YOUR CODE]]
        return result

    def trape(self):
        top, bottom, hight = map(int, input('윗변, 아래변, 높이를 입력하세요. 예시 : 윗변,아랫변,높이\n >>>').split(','))
        area = (top + bottom)* hight/2
        result = '사다리꼴의 넓이는 : ' + str(area)
        return result

a = Square()  # 객체 생성 & 2, 3을 각각 입력해 봅시다.

if a.square == 1:
  print(a.rect())
elif a.square == 2:
  print(a.par())
elif a.square == 3:
  print(a.trape())
```

키오스크 만들기

```python
class Kiosk:
    def __init__(self):
        # 전체 메뉴 리스트
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        # 전체 가격 리스트
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]

    # 메뉴 출력 메서드
    def menu_print(self):
        # 번호와 메뉴이름, 가격 순서대로 출력
        for i in range(len(self.menu)):
            print(i + 1, self.menu[i], ' : ', self.price[i], '원')

    # 주문 메서드
    def menu_select(self):
        print()  # 줄 바꿈
        n = 0    # 기본값 0

        # 1~6 사이 번호 입력될 때까지 반복
        while n < 1 or len(self.menu) < n:
            n = int(input("음료 번호를 입력하세요 : "))

            # 메뉴판에 있는 번호일 때
            if 1 <= n and n <= len(self.menu):
                self.price_sum = self.price[n-1]  # 선택한 음료 가격 저장
                self.menu_name = self.menu[n-1]   # 선택한 음료 이름 저장
            else:
                # 없는 번호일 때 다시 입력 요청
                print("없는 메뉴입니다. 다시 주문해 주세요.")

        t = 0  # 기본값 0
        # 1(HOT) 또는 2(ICE) 입력될 때까지 반복
        while t != 1 and t != 2:
            t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
            if t == 1:
                self.temp = "HOT"   # 1 입력하면 HOT 저장
            elif t == 2:
                self.temp = "ICE"   # 2 입력하면 ICE 저장
            else:
                # 1, 2 외 입력하면 다시 요청
                print("1과 2 중 하나를 입력하세요.\n")

        # 최종 주문 결과 출력
        print('주문 음료', self.temp, self.menu_name, ' : ', self.price_sum, '원')

    # 지불
    def pay(self):
        print()  # 줄 바꿈
        # 결제 금액 출력
        print(f'결제 금액 : {self.price_sum}원')

        p = 0  # 기본값 0
        # 1(현금) 또는 2(카드) 입력될 때까지 반복
        while p != 1 and p != 2:
            try:
                p = int(input("현금은 1을, 카드는 2를 입력하세요 : "))

                if p == 1:
                    # 현금 선택하면 직원 호출
                    print("직원을 호출하겠습니다.")
                elif p == 2:
                    # 카드 선택하면 카드 삽입 안내
                    print("IC칩 방향에 맞게 카드를 꽂아주세요.")
                else:
                    # 1, 2 외 입력하면 다시 요청
                    print("다시 결제를 시도해 주세요.")
                    p = 0  # 다시 입력받기 위해 초기화

            except ValueError:
                # 숫자 아닌 값 입력하면 오류 처리
                print("다시 결제를 시도해 주세요.")

    # 주문서 출력
    def table(self):
       # 위쪽 테두리 출력
        print('⟝' + '-' * 30 + '⟞')

        # 위쪽 빈 공간
        for i in range(3):
            print('|' + ' ' * 31 + '|')

        # 주문 내역 출력 (온도 + 메뉴이름 + 가격)
        print('|', self.temp, self.menu_name, ' : ', self.price_sum, '원', '|')

        # 합계 금액 출력
        print('|', '합계 금액 :', self.price_sum, '원', ' ' * 10, '|')

        # 아래쪽 빈 공간
        for i in range(3):
            print('|' + ' ' * 31 + '|')

        # 아래쪽 테두리 출력
        print('⟝' + '-' * 30 + '⟞')

a = Kiosk()  # 객체 생성
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력


#셀실행이 안됩니다 계속 대기상태여서 실행은 못했습니다 이코드만 제출합니다
```
