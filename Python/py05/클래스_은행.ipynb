은행계좌 클레스

```python
import random

class Account:

    # Q2. 클래스 변수 - 계좌 개수 저장
    account_count = 0

    # Q1. 생성자 - 예금주, 초기잔액 입력받기
    def __init__(self, owner, balance):
        # 힌트: 은행이름은 SC은행으로 고정
        self.bank = "sc bank"
        # 힌트: 예금주 저장
        self.owner = owner
        # 힌트: 초기 잔액 저장
        self.balance = balance
        # 힌트: 계좌번호 랜덤 생성 (3자리-2자리-6자리)
        # random.randint() 사용!
        self.account_num = f"{random.randint(100,999)}-{random.randint(10,99)}-{random.randint(100000,999999)}"
        # 힌트: 계좌 개수 +1
        Account.account_count += +1
        # Q7. 입금 횟수 기본값 0
        self.deposit_count = 0
        # Q10. 입금/출금 내역 리스트
        self.deposit_list = []
        self.withdraw_list = []

    # Q3. 계좌 개수 출력 메서드
    def get_account_num(self):
        # 힌트: Account.account_count 출력
        print(f"계좌 개수 : {Account.account_count}개")

    # Q4. 입금 메서드 (최소 1원 이상만 가능)
    def deposit(self, amount):
        # 힌트: amount 가 1원 이상이면 잔고에 추가
        if amount >= 1:
            self.balance += amount
            # 힌트: 입금 횟수 +1
            self.deposit_count += 1
            # Q10. 입금 내역 리스트에 추가
            self.deposit_list.append(amount)
            # Q7. 입금 횟수가 5회면 이자 1% 지급
            if self.deposit_count % 5 == 0:
                interest = int(self.balance * 0.01)
                self.balance += interest
                print(f"이자 {interest}원 지급!")
        else:
            # 힌트: 1원 미만이면 오류 메시지 출력
            print("입금은 최소 1원 이상만 가능합니다")

    # Q5. 출금 메서드 (잔고 이상 출금 불가)
    def withdraw(self, amount):
        # 힌트: amount 가 잔고 이하면 출금 가능
        if amount <= self.balance :
            self.balance -= amount
            # Q10. 출금 내역 리스트에 추가
            self.withdraw_list.append(amount)
        else:
            # 힌트: 잔고 부족 메시지 출력
            print("잔고가 부족합니다")

    # Q6. 정보 출력 메서드 (잔고 세자리마다 쉼표)
    def display_info(self):
        # 힌트: f스트링 + {:,} 로 세자리 쉼표!
        print(f"은행이름: {self.bank}")
        print(f"예금주: {self.owner}")
        print(f"계좌번호: {self.account_num}")
        print(f"잔고: {self.balance:,}원")  # {:,} = 세자리 쉼표!

    # Q10. 입금 내역 출력 메서드
    def deposit_history(self):
        print("입금 내역:")
        for i in self.deposit_list:
            print(f"{i}원")

    # Q10. 출금 내역 출력 메서드
    def withdraw_history(self):
        print("출금 내역")
        for i in self.withdraw_list:
            print(f"{i}")


# Q8. 3개 이상 인스턴스 생성 후 리스트에 저장
accounts = []
accounts.append(Account("김성훈", 1000000))  # 힌트: 예금주, 초기잔액
accounts.append(Account("유정욱", 10000))
accounts.append(Account("홍길동", 5000000))

# Q9. 잔고 100만원 이상인 고객만 출력
for account in accounts:
    if account.balance >= 1000000:
        account.display_info()
```
